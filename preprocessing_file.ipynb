#### Import packages   
Packages such pandas, regex

```python
import pandas as pd
import re
```

```python
df = pd.read_csv('unstructured.csv')
df
```

#### Remove the NaN values from the fetch response

```python
new_df = df.copy()
new_df = new_df[~new_df.fetch_response.isnull()]
new_df
```

```python
def integer_2_text(df, x):
    row = df[df.Unique_id == int(x)]
    return '' if (len(row) == 0) else row.iloc[0]['Text']
```

```python
new_df.loc[:, 'fetch_response'] = new_df.fetch_response.apply(
    lambda x: integer_2_text(df, x))
```

```python
new_df
```

```python
new_df.rename(columns={'Unique_id': 'Unique_id', 'Text': 'answer', 'fetch_response': 'question'},
                           inplace=True)
new_df = new_df[['Unique_id', 'question', 'answer']]
```

```python
new_df
```

```python
def clean_text(data):
    data = ' '.join(word for word in data.split(' ') if not word.startswith('@'))
    data = re.sub(r'\d+$', '', data)
    return data
```

```python
new_df.loc[:, 'question'] = new_df.question.apply(clean_text)
new_df.loc[:, 'answer'] = new_df.answer.apply(clean_text)
```

```python
new_df
```

```python
new_df.to_csv('processed.csv')
```
